```python
from fbprophet import Prophet
import pandas as pd
```

```python
data_sample = pd.read_csv('data.csv',',')
```

# Оценка отдельно тенденций каждого из хостов 

```python
All_data_forecast = pd.DataFrame()
```

```python
data_sample_iOS = data_sample[data_sample["host"] == 'iOS']
data_sample_web = data_sample[data_sample["host"] == 'web']
data_sample_Android = data_sample[data_sample["host"] == 'Android']
data_sample_mobile_web = data_sample[data_sample["host"] == 'mobile_web']
```

```python
predictions = 30
```

```python
data_sample_iOS.rename(columns = {'date':'ds', 'host':'host',
                              'searches':'y'}, inplace = True)
data_sample_web.rename(columns = {'date':'ds', 'host':'host',
                              'searches':'y'}, inplace = True)
data_sample_Android.rename(columns = {'date':'ds', 'host':'host',
                              'searches':'y'}, inplace = True)
data_sample_mobile_web.rename(columns = {'date':'ds', 'host':'host',
                              'searches':'y'}, inplace = True)
```

```python
data_sample_iOS= data_sample_iOS[['ds', 'y']]
data_sample_web= data_sample_web[['ds', 'y']]
data_sample_Android= data_sample_Android[['ds', 'y']]
data_sample_mobile_web= data_sample_mobile_web[['ds', 'y']]
```

```python
m_iOS = Prophet()
m_iOS.fit(data_sample_iOS)

m_web = Prophet()
m_web.fit(data_sample_web)

m_Android = Prophet()
m_Android.fit(data_sample_Android)

m_mobile_web = Prophet()
m_mobile_web.fit(data_sample_mobile_web)
```

```python
future_iOS = m_iOS.make_future_dataframe(periods=365)
forecast_iOS = m_iOS.predict(future_iOS)

future_web = m_web.make_future_dataframe(periods=365)
forecast_web = m_web.predict(future_web)

future_Android = m_Android.make_future_dataframe(periods=365)
forecast_Android = m_Android.predict(future_Android)

future_mobile_web = m_mobile_web.make_future_dataframe(periods=365)
forecast_mobile_web = m_mobile_web.predict(future_mobile_web)
```

```python
m_iOS.plot(forecast_iOS)
m_Android.plot(forecast_Android)
m_web.plot(forecast_web)
m_mobile_web.plot(forecast_mobile_web)
```

```python
forecast_iOS[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
```

```python
All_data_forecast['ds'] = forecast_iOS['ds']
All_data_forecast = All_data_forecast.merge(forecast_iOS[['ds', 'yhat', 'yhat_lower', 'yhat_upper']], on='ds')

All_data_forecast.rename(columns = {'ds':'ds', 
                                  'yhat':'iOS',
                                  'yhat_lower':'iOS_pi_lower',
                                  'yhat_upper':'iOS_pi_upper'}, inplace = True)
```
